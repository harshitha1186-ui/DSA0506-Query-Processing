6.Write a Pandas program to create a scatter plot of the trading
volume/stock prices of Alphabet Inc. stock between two specific dates.

```python
import pandas as pd
import matplotlib.pyplot as plt
alphabet_stock_data = pd.DataFrame({
    'Date': pd.date_range(start='2022-01-01', periods=6),
    'Close': [2800, 2825, 2790, 2810, 2830, 2850],
    'Volume': [1500000, 1800000, 1600000, 1700000, 1750000, 1900000]
})
alphabet_stock_data['Date'] = pd.to_datetime(alphabet_stock_data['Date'])
start_date = '2022-01-01'
end_date = '2022-01-06'
filtered_data = alphabet_stock_data[
    (alphabet_stock_data['Date'] >= start_date) &
    (alphabet_stock_data['Date'] <= end_date)
]
plt.figure(figsize=(8, 6))
plt.scatter(filtered_data['Volume'], filtered_data['Close'])
plt.xlabel('Trading Volume')
plt.ylabel('Stock Price (Close)')
plt.title('Alphabet Inc. Stock Price vs Trading Volume')
plt.show()
```

14.Write a Pandas program to find and replace the missing values in a given
DataFrame which do not have any valuable information.

```python
import pandas as pd
import numpy as np
data = {
    'Name': ['Alice', 'Bob', np.nan, 'David'],
    'Age': [25, np.nan, np.nan, 40],
    'City': [np.nan, 'New York', 'Los Angeles', np.nan],
    'Score': [np.nan, 88, 92, np.nan]
}
df = pd.DataFrame(data)
print("Original DataFrame:\n", df)
df['Name'] = df['Name'].fillna('Unknown')
df['City'] = df['City'].fillna('Unknown')
df['Age'] = df['Age'].fillna(0)
df['Score'] = df['Score'].fillna(0)
print("\nDataFrame after replacing missing values:\n", df)
```

23.Write a Python program to draw a line using given axis values taken from a
text file, with suitable label in the x axis, y axis and a title.
Test Data:
test.txt
1 2
2 4
3 1

```python
import matplotlib.pyplot as plt
with open("test.txt", "w") as f:
    f.write("1 2\n")
    f.write("2 4\n")
    f.write("3 6\n")
    f.write("4 8\n")
    f.write("5 10\n")

x = []
y = []

with open("test.txt") as f:
    for line in f:
        a, b = map(int, line.split())
        x.append(a)
        y.append(b)

plt.plot(x, y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Line Graph from Text File")
plt.show()
```

31.Write a Python program to create a stacked bar plot with error bars.
Note: Use bottom to stack the women?s bars on top of the men?s bars.
Sample Data:
Means (men) = (22, 30, 35, 35, 26)
Means (women) = (25, 32, 30, 35, 29)
Men Standard deviation = (4, 3, 4, 1, 5)
Women Standard deviation = (3, 5, 2, 3, 3)

```python
import matplotlib.pyplot as plt
import numpy as np
men_means = (22, 30, 35, 35, 26)
women_means = (25, 32, 30, 35, 29)
men_std = (4, 3, 4, 1, 5)
women_std = (3, 5, 2, 3, 3)
x = np.arange(len(men_means))
plt.bar(x, men_means, yerr=men_std)
plt.bar(x, women_means, bottom=men_means, yerr=women_std)
plt.xlabel("Group")
plt.ylabel("Scores")
plt.title("Stacked Bar Plot with Error Bars")
plt.show()
```
